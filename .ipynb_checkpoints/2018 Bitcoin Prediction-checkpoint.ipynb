```python
import numpy as np
import csv
import time
import random
import matplotlib.pylab as plt
import pandas as pd
```

```python
df = pd.read_csv('Bitcoin.csv')
df['Date'] = pd.to_datetime(df.Date)
df = df.sort_values(by='Date')
train = df[0:1500]
test = df[1500:]
```

```python
ax = train.plot(x = 'Date', y = 'Close', figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = '2014-2018', color = "black")
test.plot(ax=ax, x = 'Date', y = 'Close' ,figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = '2018-2019', color = "orange")
plt.show()
```

```python
dd = np.asarray(train.Close)
y_hat = test.copy()
y_hat['naive'] = dd[len(dd)-1]
ax = train.plot(x = 'Date', y = 'Close', figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = '2014-2018', color = "black")
bx = test.plot(ax=ax, x = 'Date', y = 'Close' ,figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = '2018-2019', color = "orange")
y_hat.plot(ax=bx, x = 'Date', y = 'naive', figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = 'Predicted', color = "blue")
plt.show()
```

```python
dd = np.asarray(train.Close)
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['Close'].mean()
ax = train.plot(x = 'Date', y = 'Close', figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = '2014-2018', color = "black")
bx = test.plot(ax=ax, x = 'Date', y = 'Close' ,figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = '2018-2019', color = "orange")
y_hat_avg.plot(ax=bx, x = 'Date', y = 'avg_forecast', figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = 'Predicted', color = "blue")
plt.show()
```

```python
dd = np.asarray(train.Close)
y_hat_avg = test.copy()
y_hat_avg['moving_avg_forecast'] = train['Close'].rolling(260).mean().iloc[-1]
ax = train.plot(x = 'Date', y = 'Close', figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = '2014-2018', color = "black")
bx = test.plot(ax=ax, x = 'Date', y = 'Close' ,figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = '2018-2019', color = "orange")
y_hat_avg.plot(ax=bx, x = 'Date', y = 'moving_avg_forecast', figsize=(15,8), title= 'Bitcoin Price', fontsize=14, label = 'Predicted', color = "blue")
plt.show()
```
